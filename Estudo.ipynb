```
def calcular_imc(peso, altura):
    return peso / (altura ** 2)


def classificar_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif imc < 25:
        return "Peso ideal"
    elif imc < 30:
        return "Sobrepeso"
    elif imc < 35:
        return "Obesidade grau I"
    elif imc < 40:
        return "Obesidade grau II"
    else:
        return "Obesidade mórbida"


def main():
    print("=== CALCULADORA DE IMC ===")


    try:
        peso = float(input("Digite seu peso (ex: 70): "))
        altura = float(input("Digite sua altura (ex: 1.70): "))


        if altura > 3:
          altura = altura / 100


        if peso <= 0 or altura <= 0:
            print("Peso e altura devem ser maiores que zero.")
            return


        imc = calcular_imc(peso, altura)
        classificacao = classificar_imc(imc)


        print(f"\nSeu peso: {peso}")
        print(f"Sua altura: {altura:.2f}")
        print(f"Seu IMC é: {imc:.2f}")
        print(f"Classificação: {classificacao}")

    except ValueError:
        print("Por favor, digite apenas números válidos.")


main()
```

```
def celsius_para_fahrenheit(celsius):
    return (celsius * 9/5) + 32


def fahrenheit_para_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9


def main():
    print("=== CONVERSÃO DE TEMPERATURA ===")

    try:
        opcao = input("Digite C para converter Celsius ou F para converter Fahrenheit: ").upper()

        if opcao == "C":
            celsius = float(input("Digite a temperatura em Celsius: "))
            resultado = celsius_para_fahrenheit(celsius)
            print(f"{celsius}°C equivale a {resultado:.2f}°F")

        elif opcao == "F":
            fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
            resultado = fahrenheit_para_celsius(fahrenheit)
            print(f"{fahrenheit}°F equivale a {resultado:.2f}°C")

        else:
            print("Opção inválida.")

    except ValueError:
        print("Por favor, digite apenas números válidos.")


main()
```
